```python
# Problema 1: Función que calcule la molaridad de una solución
# Fórmula general: M = molesSoluto / volumenL
# Parámetros: masa en gramos, peso molecular (g/mol), volumen en mL

def calcularMolaridad(masaSoluto, pesoMolecular, volumenMl):
    molesSoluto = masaSoluto / pesoMolecular
    volumenL = volumenMl / 1000
    M = molesSoluto / volumenL
    return M

# Ejemplo #1: Cálculo de la molaridad de KCl
molaridad = calcularMolaridad(20, 74.55, 1000)
print(f"Molaridad de la solución de KCl: {molaridad} mol/L")
```

```python
import math

# Problema 2: Función que calcule el pH de una solución de un ácido fuerte
# Fórmula general: pH = -log10([H+])
# Parámetro: concentración de iones de hidrógeno en mol/L.

def calcularPh(concentracionH):
    return -math.log10(concentracionH)

# Ejemplo #2: Cálculo del pH de una solución de ácido fuerte con concentración 0.001 M
ph = calcularPh(0.001)
print(f"pH de la solución: {ph}")

```

```python
# Problema 3: Función que calcula la masa molar de un gas ideal
# Fórmula general: M = (masa * R * T) / (P * V)
# Parámetros: masa en gramos, volumen en mL, temperatura en °C, presión en atm.

def calcularMasaMolar(masa, volumenMl, temperaturaC, presionAtm):
    volumenL = volumenMl / 1000  # Convertir volumen de mL a L
    temperaturaK = temperaturaC + 273.15  # Convertir temperatura de °C a K
    R = 0.0821  # Constante de los gases en L·atm/(mol·K)
    M = (masa * R * temperaturaK) / (presionAtm * volumenL)
    return M

# Ejemplo #3: Cálculo de la masa molar para un gas ideal con masa 5 g, volumen 500 mL, temperatura 50°C, y presión 1 atm
masaMolar = calcularMasaMolar(5, 500, 50, 1)
print(f"Masa molar del gas: {masaMolar} g/mol")

```

```python
# Problema 4: Función que calcula la concentración del ácido y de la base conjugada para un buffer
# Fórmula general: [Base] = 10^(pH - pKa) * [Ácido]
# Parámetros: pH deseado, constante de acidez (Ka) y volumen en mL

def calcularConcentracionesBuffer(phDeseado, ka, volumenMl):
    concentracionAcido = 10 ** (phDeseado - math.log10(ka))
    concentracionBase = concentracionAcido * (10 ** (phDeseado - math.log10(ka)))
    volumenL = volumenMl / 1000
    return concentracionAcido * volumenL, concentracionBase * volumenL

# Ejemplo #4: Cálculo de concentraciones para un buffer con pH deseado de 5.5 y Ka de 1.8e-5
concentracionAcido, concentracionBase = calcularConcentracionesBuffer(5.5, 1.8e-5, 600)
print(f"Concentración de ácido: {concentracionAcido} mol/L")
print(f"Concentración de base: {concentracionBase} mol/L")

```

```python
import math

# Problema 4: Función que calcula la concentración del ácido y base conjugada para un buffer
# usando la ecuación de Henderson-Hasselbalch.
# Parámetros: pH deseado, Ka (constante de acidez), volumen en mL

def calcularConcentracionesBuffer(pH_deseado, Ka, volumen_ml, concentracion_total=0.5):
    # Convertir Ka a pKa
    pKa = -math.log10(Ka)

    # Calcular la relación [Base] / [Ácido]
    relacion_base_acido = 10 ** (pH_deseado - pKa)

    # Determinar las concentraciones de ácido y base
    concentracion_acido = concentracion_total / (1 + relacion_base_acido)
    concentracion_base = concentracion_acido * relacion_base_acido

    # Convertir a moles ajustados al volumen
    moles_acido = concentracion_acido * (volumen_ml / 1000)
    moles_base = concentracion_base * (volumen_ml / 1000)

    return concentracion_acido, concentracion_base, moles_acido, moles_base

# Ejemplo #4: Buffer con pH 7.2, Ka = 6.2e-8 (ácido fosfórico), volumen 200 mL
concentracion_acido, concentracion_base, moles_acido, moles_base = calcularConcentracionesBuffer(7.2, 6.2e-8, 200)
print(f"Concentración de ácido: {concentracion_acido} M")
print(f"Concentración de base conjugada: {concentracion_base} M")
print(f"Moles de ácido: {moles_acido} mol")
print(f"Moles de base conjugada: {moles_base} mol")
```

```python
# Problema 5: Función que calcula la energía liberada en una reacción química
# Fórmula general: Energía = molesReactivo * deltaHFormacion
# Parámetros: moles de reactivo y ΔH de formación (kJ/mol)

def calcularEnergiaLiberada(molesReactivo, deltaHFormacion):
    energia = molesReactivo * deltaHFormacion
    return energia

# Ejemplo #5: Cálculo de energía liberada con 3 moles de reactivo y ΔH de formación de -150.5 kJ/mol
energiaLiberada2 = calcularEnergiaLiberada(3, -150.5)
print(f"Energía liberada en la reacción: {energiaLiberada2} kJ")

```

```python
# Problema 6: Función que calcula el rendimiento porcentual de una reacción
# Fórmula general: Rendimiento % = (masaExperimental / masaTeorica) * 100
# Parámetros: masa teórica y masa experimental (en gramos)

def calcularRendimientoPorcentual(masaTeorica, masaExperimental):
    rendimiento = (masaExperimental / masaTeorica) * 100
    return rendimiento

# Ejemplo #6: Cálculo de rendimiento porcentual con una masa teórica de 20 g y una masa experimental de 18 g
rendimiento = calcularRendimientoPorcentual(20, 18)
print(f"Rendimiento porcentual de la reacción: {rendimiento}%")
```

```python
# Problema 7: Función que calcula el calor específico de un metal
# Fórmula general: Calor específico = calorAbsorbido / (masa * deltaT)
# Parámetros: masa en gramos, cambio de temperatura (°C) y calor absorbido (J)

def calcularCalorEspecifico(masa, deltaT, calorAbsorbido):
    calorEspecifico = calorAbsorbido / (masa * deltaT)
    return calorEspecifico

# Ejemplo #7: Cálculo del calor específico con masa de 30 g, ΔT de 15 °C, y calor absorbido de 300 J
calorEspecifico = calcularCalorEspecifico(30, 15, 300)
print(f"Calor específico del metal: {calorEspecifico} J/g°C")

```

```python
# Problema 8: Función que calcula la presión osmótica de una disolución
# Fórmula general: Presión osmótica = Molaridad * R * T
# Parámetros: concentración en gramos por litro, peso molecular del soluto y temperatura en °C

def calcularPresionOsmotica(concentracion_gL, pesoMolecular, temperaturaC):
    temperaturaK = temperaturaC + 273.15  # Convertir temperatura a Kelvin
    molaridad = concentracion_gL / pesoMolecular  # Calcular molaridad (moles/L)
    R = 0.0821  # Constante de los gases en L·atm/(mol·K)
    presionOsmotica = molaridad * R * temperaturaK
    return presionOsmotica

# Ejemplo #8: Cálculo de la presión osmótica para una disolución con 2 g/L de soluto de peso molecular 58.44 g/mol a 30°C
presionOsmotica2 = calcularPresionOsmotica(2, 58.44, 30)
print(f"Presión osmótica de la disolución: {presionOsmotica2} atm")

```

```python
# Problema 9: Función que calcula el producto de solubilidad (Kps) de una sal insoluble
# Fórmula general para PbCl2: Kps = 4 * s^3
# Parámetro: solubilidad en mol/L

def calcularKps(solubilidad, compuesto):
    if compuesto == "PbCl2":
        kps = 4 * (solubilidad ** 3)  # Para PbCl2, Kps = 4s^3
    else:
        raise ValueError("Compuesto no soportado. Usa 'PbCl2'.")
    return kps

# Ejemplo #9: Cálculo de Kps para PbCl2 con una solubilidad de 0.02 mol/L
kps = calcularKps(0.02, "PbCl2")
print(f"Kps de PbCl2: {kps}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Problema 10: Cálculo de la velocidad de reacción en función de la concentración de reactivo
# Fórmula general para una reacción de primer orden: v = k * [A]
# Parámetros: constante de velocidad (k) y array de concentraciones [A]

def calcularVelocidadReaccion(k, concentraciones):
    velocidad = k * concentraciones  # Calcula la velocidad de reacción
    return velocidad

# Ejemplo #10: Gráfica de la velocidad de reacción para una reacción de primer orden con k = 0.02 s^-1
concentraciones = np.linspace(0.1, 1.0, 10)  # Concentraciones de 0.1 a 1.0 mol/L
velocidades = calcularVelocidadReaccion(0.02, concentraciones)

# Graficar concentración vs. velocidad de reacción
plt.plot(concentraciones, velocidades, marker='o')
plt.xlabel("Concentración [A] (mol/L)")
plt.ylabel("Velocidad de reacción (mol/L·s)")
plt.title("Velocidad de reacción en función de la concentración")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Problema 10.1: Cálculo de la velocidad de reacción de descomposición de H₂O₂
# Fórmula general para una reacción de primer orden: v = k * [H₂O₂]
# Parámetros: constante de velocidad (k) y array de concentraciones de H₂O₂

def calcularVelocidadReaccion(k, concentraciones):
    velocidad = k * concentraciones  # Calcula la velocidad de reacción
    return velocidad

# Ejemplo #10: Gráfica de la velocidad de descomposición de H₂O₂ con k = 0.02 s^-1
concentraciones = np.linspace(0.1, 1.0, 10)  # Concentraciones de H₂O₂ de 0.1 a 1.0 mol/L
velocidades = calcularVelocidadReaccion(0.02, concentraciones)

# Graficar concentración vs. velocidad de reacción
plt.plot(concentraciones, velocidades, marker='o')
plt.xlabel("Concentración de H₂O₂ (mol/L)")
plt.ylabel("Velocidad de reacción (mol/L·s)")
plt.title("Velocidad de descomposición de H₂O₂ en función de la concentración")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Problema 10.2: Cálculo de la concentración de H₂O₂ en función del tiempo para una reacción de primer orden
# Fórmula general para una reacción de primer orden: [A] = [A_0] * exp(-k * t)
# Parámetros: concentración inicial (A_0), constante de velocidad (k) y array de tiempos

def calcularConcentracionEnTiempo(A_0, k, tiempos):
    concentraciones = A_0 * np.exp(-k * tiempos)  # Calcula la concentración en función del tiempo
    return concentraciones

# Ejemplo #10: Gráfica de la concentración de H₂O₂ en función del tiempo para k = 0.02 s^-1
A_0 = 1.0  # Concentración inicial de H₂O₂ en mol/L
k = 0.02  # Constante de velocidad en s^-1
tiempos = np.linspace(0, 200, 50)  # Tiempo en segundos

# Calcula la concentración de H₂O₂ en función del tiempo
concentraciones = calcularConcentracionEnTiempo(A_0, k, tiempos)

# Graficar tiempo vs. concentración de H₂O₂
plt.plot(tiempos, concentraciones, marker='o')
plt.xlabel("Tiempo (s)")
plt.ylabel("Concentración de H₂O₂ (mol/L)")
plt.title("Descomposición de H₂O₂ en función del tiempo")
plt.grid()
plt.show()
```
